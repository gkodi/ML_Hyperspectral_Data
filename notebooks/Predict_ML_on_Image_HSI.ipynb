```python
def predict_image(path_to_image, path_to_model):
    img = open_image(path_to_image)

    y = img[:,:,:].shape[0] # y axis
    x = img[:,:,:].shape[1] # x axis
    z = img[:,:,:].shape[2] # z axis (bands)

    a = img[:,:,:].reshape(x*y,z)  #Reshape to 2D array

    df = pd.DataFrame(a, columns=[f"band_{i+1}" for i in range(z)])

    with open(path_to_model, "rb") as f:
      loaded_model = pickle.load(f) #load the model
    predictions = loaded_model.predict(a)

    for i, label in enumerate(loaded_model.labels):
        df[label] = predictions[:, i]

    return df
```
